```python
import matplotlib.pyplot as plt
import pandas as pd
import shutil

# Load the data
log_path = "model_logs/v17.1.csv"
df = pd.read_csv(log_path)

# Helper function to plot and save
def plot_metric(x, y, title, ylabel, filename, window=20):
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label="Raw", alpha=0.3, color='gray')
    plt.plot(x, y.rolling(window=window).mean(), label=f"Moving Avg (n={window})", color='blue', linewidth=2)
    plt.xlabel("Training Iterations")
    plt.ylabel(ylabel)
    plt.title(f"{title} (v17.1)")
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Save local and backup
    plt.savefig(filename)
    shutil.copy(filename, f"/content/drive/MyDrive/colab_package_backup/{filename}")
    plt.show()
    print(f"Saved {filename} to Drive")

# 1. Reward Plot
plot_metric(df["iteration"], df["reward_mean"], "Training Reward", "Average Reward", "reward_plot.png")

# 2. Length Plot
plot_metric(df["iteration"], df["length_mean"], "Episode Length", "Average Length", "length_plot.png")

# 3. Loss Plot
plot_metric(df["iteration"], df["loss"], "Training Loss", "Loss", "loss_plot.png")
```

```python
%cd /content/colab_package

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

from agent import DeepQLearningAgent
from game_environment import SnakeNumpy

# 1) Env + agent match the trained model: 2 frames, 4 actions, v17.1
env = SnakeNumpy(board_size=10, frames=2, games=1,
                 obstacles=True, version='v17.1')
n_actions = env.get_num_actions()

agent = DeepQLearningAgent(
    board_size=10,
    frames=2,
    n_actions=n_actions,
    use_target_net=True,
    version='v17.1'
)
agent.load_model(file_path='models/v17.1', iteration=200000)

# Try to get the final loss from the logs if available
try:
    final_loss = df_logs['loss'].iloc[-1]
except:
    final_loss = 0.0

# 2) Play one game and store board snapshots
state = env.reset(stateful=True)
done = np.array([0], dtype=np.uint8)

# We'll store tuples of (board, title_info)
snapshots = []
steps = 0

# Initial state info
current_info = {
    'time': 0,
    'food': 0,
    'length': 2,
    'reward': 0
}

while not done.all() and steps < 40:   # limit to 40 steps
    # Capture current board
    board_img = env._board[0].copy()

    # Prepare title for this frame
    title_str = (
        f"Time: {current_info['time']} | Score: {current_info['food']}\n"
        f"Reward: {current_info['reward']} | Len: {current_info['length']}\n"
        f"Model Loss: {final_loss:.4f}"
    )
    snapshots.append((board_img, title_str))

    legal_moves = env.get_legal_moves()
    action = agent.move(state, legal_moves, env.get_values())
    state, reward, done, info, next_legal = env.step(action)

    # Update info for next frame (or current if we captured post-step, but pre-step capture is better for 'Step 0')
    # Note: 'info' contains arrays for batch, extract index 0
    current_info = {
        'time': info['time'][0],
        'food': info['food'][0],
        'length': info['length'][0],
        'reward': reward[0]
    }

    steps += 1

# 3) Plot 4 snapshots from the game with custom colors
num_frames = min(4, len(snapshots))
fig, axes = plt.subplots(1, num_frames, figsize=(4*num_frames, 4))

if num_frames == 1:
    axes = [axes]

# Custom Color Palette
# 0: Empty -> Grey (#D3D3D3)
# 1: Body -> Green (#008800)
# 2: Head -> Light Green (#7CFC00)
# 3: Food -> Red (#FF0000)
# 4: Border -> Dark Grey (#505050)
cmap = ListedColormap(["#D3D3D3", "#008800", "#7CFC00", "#FF0000", "#505050"])

# Select 4 evenly spaced frames if we have many, or just the first 4
indices = np.linspace(0, len(snapshots)-1, num_frames, dtype=int)

for i, idx in enumerate(indices):
    board, title = snapshots[idx]
    board = board.squeeze()               # (10, 10)
    vmin, vmax = 0, 4

    axes[i].imshow(board, cmap=cmap, vmin=vmin, vmax=vmax)
    axes[i].set_title(title, fontsize=10)
    axes[i].set_xticks([])
    axes[i].set_yticks([])

plt.tight_layout()
plt.show()
```
